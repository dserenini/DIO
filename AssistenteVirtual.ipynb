```python
from gtts import gTTS
from IPython.display import Audio
import speech_recognition as sr
import wikipedia
import webbrowser  # Substitui o pywhatkit
import os

# Função para converter texto em áudio
def text_to_speech(text, lang='pt'):
    tts = gTTS(text=text, lang=lang, slow=False)
    tts.save("output.mp3")
    display(Audio("output.mp3", autoplay=True))
    os.remove("output.mp3")

# Função para reconhecer fala a partir de um arquivo de áudio
def speech_to_text(audio_file):
    recognizer = sr.Recognizer()
    with sr.AudioFile(audio_file) as source:
        audio = recognizer.record(source)
        try:
            text = recognizer.recognize_google(audio, language='pt-BR')
            print(f"Texto reconhecido: {text}")
            return text
        except sr.UnknownValueError:
            print("Não entendi o áudio.")
            return None
        except sr.RequestError:
            print("Erro ao conectar ao serviço de reconhecimento de fala.")
            return None

# Função para executar comandos
def executar_comando(comando):
    if 'wikipedia' in comando:
        texto_pesquisa = comando.replace('wikipedia', '').strip()
        try:
            resultado = wikipedia.summary(texto_pesquisa, sentences=2)
            print(resultado)
            text_to_speech(resultado)
        except wikipedia.exceptions.DisambiguationError as e:
            text_to_speech("Há muitas opções. Por favor, seja mais específico.")
        except wikipedia.exceptions.PageError:
            text_to_speech("Desculpe, não encontrei nada sobre isso na Wikipedia.")

    elif 'abrir youtube' in comando:
        text_to_speech("Abrindo o YouTube.")
        webbrowser.open("https://www.youtube.com")  # Abre o YouTube no navegador

    elif 'farmácia mais próxima' in comando:
        text_to_speech("Aqui está a localização da farmácia mais próxima.")
        webbrowser.open("https://www.google.com/maps/search/farmácia+próxima")  # Abre o Google Maps

    else:
        text_to_speech("Desculpe, não entendi o comando.")

# Função principal
def main():
    # Faça o upload do arquivo de áudio no Google Colab
    audio_file = "OpenYoutube.wav"  # Substitua pelo nome do seu arquivo de áudio

    # Reconhece o texto do áudio
    comando = speech_to_text(audio_file)

    # Executa o comando reconhecido
    if comando:
        executar_comando(comando.lower())

# Executa o sistema
main()
```

```python
#Caso seja utilizado com microfone, poderia fazer um whileTrue aqui

while True:
    print("Aguardando comando...")
    comando = speech_to_text()
    if comando:
        if 'parar' in comando.lower():
            text_to_speech("Encerrando o sistema.")
            break
        executar_comando(comando.lower())
```
